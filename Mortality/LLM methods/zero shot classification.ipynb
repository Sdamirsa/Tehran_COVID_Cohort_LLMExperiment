```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification",
                      model="facebook/bart-large-mnli")
```

```python
!pip install -q datasets
```

```python
from datasets import load_dataset
dataset = load_dataset("Moreza009/Internal_validation")
```

```python
train_ds = dataset["test"].shuffle(seed=42).select(range(100))
```

```python
train_ds
```

```python
train_ds['patient medical hidtory']
```

```python
list_results = []

for i in train_ds['patient medical hidtory'] :
  sequence_to_classify = i
  candidate_labels = ['survives', 'dies']
  classing=classifier(sequence_to_classify, candidate_labels)
  scores =classing['scores']
  highest_value_position = scores.index(max(scores))

  result = classing['labels'][highest_value_position]
  list_results.append(result)

print(list_results)
```
