```python
import pandas as pd
import numpy as np

# Set the seed for reproducibility
np.random.seed(42)

# Define the number of rows and columns
num_rows = 10
num_columns = 5

# Create a random DataFrame with a mix of data types
data = {
    'Column1': np.random.choice([None, 'A', 'B', 'C', 1, 2, 3, 4], size=num_rows),
    'Column2': np.random.choice([None, 'X', 'Y', 'Z', 0.1, 0.2, 0.3, 0.4], size=num_rows),
    'Column3': np.random.choice([None, 'D', 'E', 'F', 5, 6, 7, 8], size=num_rows),
    'Column4': np.random.choice([None, 'G', 'H', 'I', 9, 10, 11, 12], size=num_rows),
    'Column5': np.random.choice([None, 'J', 'K', 'L', 13, 14, 15, 16], size=num_rows)
}

# Create a DataFrame
df = pd.DataFrame(data)

# Replace some values with NaN to achieve 40% NaN values
df = df.mask(np.random.random(df.shape) < 0.4)

# Print the DataFrame
print(df)
```

```python
df
```

```python
# Calculate the threshold for dropping rows (90% NaN)
threshold = 0.9 * df.shape[1]

# Drop rows with more than 90% NaN values
df_cleaned = df.dropna(thresh=threshold)

# Print the cleaned DataFrame
print(df_cleaned)
```
